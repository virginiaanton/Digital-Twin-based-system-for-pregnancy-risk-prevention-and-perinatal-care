### COMPROBACIÓN RANGO VÁLIDO DE LOS DATOS

```python
def verifica_rango(dataframe):
    """
    Función que comprueba si los valores se encuentran en el rango fisiológico compatible con la vida humana
    devuelve una serie de comentarios diciendo que valores se encuentran dentro del rango y cuáles no. 
    """
    
    rangos_limites = {
        'SBP': (70,200),
        'DBP':(40,120),
        'Glucose':(70.0,220.0),
        'HeartRate':(30,300),
        'BMI':(18.5, 30)
    }
    
    results= {}
    for col, (min_val, max_val) in rangos_limites.items():
        if col in dataframe.columns:
            
            validos = dataframe[(dataframe[col] >= min_val) & (dataframe[col] <= max_val)]
            fuera_rango = len(dataframe) - len(validos)
            results[col] = {
                'val_fuera_rango':fuera_rango,
                'total_filas': len(dataframe), 
                'valores_validos': len(validos)
            }

    for col, info in results.items():
        print(f"- Valores fuera del rango de {col}:{info['val_fuera_rango']}\n")
        print(f"- Total valores de {col}:{info['total_filas']}\n")
        print(f"- Valores validos de {col}: {info['valores_validos']}\n")
    
```

```python
verifica_rango(known_age)
```

```python
def ajustar_valores(df, rangos):
    """
    Función que trunca los valores que se pasan del rango al máximo o mínimo del rango fisiológico. 
    """
    for col, (min_val, max_val) in rangos.items():
        if col in df.columns:
            df[col] = df[col].apply(
                lambda x: max(min(x, max_val), min_val) if not pd.isnull(x) else x
            )
    return df
    
```

```python
   rangos_limites = {
        'SBP': (70,200),
        'DBP':(40,120),
        'Glucose':(70.0,220.0),
        'HeartRate':(30,300),
        'BMI':(18.5, 30)
    }
known_age = ajustar_valores(known_age, rangos_limites)
unknown_age = ajustar_valores(unknown_age, rangos_limites)
```

Ahora vamos a rellenarlos. 

Rellenamos con un modelo predictivo

### MANEJO DE VALORES NAN COLUMNA COLUMNA GLUCOSE

```python
from sklearn.linear_model import LinearRegression

train_data = known_age[known_age['Glucose'].notnull()]
test_data = known_age[known_age['Glucose'].isnull()]

X_train = train_data[['AGE']]
y_train = train_data['Glucose']
X_test = test_data[['AGE']]

model = LinearRegression()
model.fit(X_train, y_train)

known_age.loc[known_age['Glucose'].isnull(), 'Glucose'] = model.predict(X_test)
```

```python
print(known_age['Glucose'].isna().sum())
```

```python
### Rellenamos el resto de valores con la mediana
known_age['DBP'] = known_age['DBP'].fillna(known_age['DBP'].median())
known_age['SBP'] = known_age['SBP'].fillna(known_age['SBP'].median())
known_age['HeartRate'] = known_age['HeartRate'].fillna(known_age['HeartRate'].median())
known_age['BMI'] = known_age['BMI'].fillna(known_age['BMI'].median())
```

### CORRELACIONES ANÁLISIS ENTRE VARIABLES

```python
corr_matrix = known_age[['AGE','SBP', 'DBP', 'BMI', 'Glucose', 'HeartRate']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title("\nHeatmap de Correlaciones de los signos vitales\n")
plt.show()
```

Podemos ver que la relación de la edad con las otras variables es muy baja (menor a 0.1), parece no tener relación lineal significativa. 
Las relaciones entre las variables en general son muy débiles. 

La relación negativa entre DBP y BMI puede darse en etapas tempranas en la obesidad, resulta extraña, según la literatura, a mayor BMI se relaciona con un aumento de la presión sanguínea debido a la rigidez de las arterias. No se ha encontrado ningún paper que evidencie lo contrario y respalde este caso. 
Esto nos lleva a pensar que la muestra no es representativa o que está sesgada ya que estamos tratando sólo con 206 pacientes.

Como faltan muchos valores podemos imputar valores por KNN. 

```python
from sklearn.impute import KNNImputer

columns_to_impute = ['SBP', 'DBP', 'Glucose', 'HeartRate', 'BMI']

imputer = KNNImputer(n_neighbors=5, weights="uniform")
imputed_data = imputer.fit_transform(unknown_age[columns_to_impute])

unknown_age[columns_to_impute] = imputed_data
print(unknown_age[columns_to_impute].isna().sum())
```
