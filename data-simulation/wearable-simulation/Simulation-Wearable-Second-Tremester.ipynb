## SIMULACIÓN DE CAMBIOS EN LOS SIGNOS VITALES DURANTE EL SEGUNDO TRIMESTRE

En esta sección vamos a simular los cambios en los signos vitales del primer trimestre al segundo. Tras hacer un poco de investigación. Vamos a simular las siguientes tendencias:
<ol>
    <li>DiastolicBP y SystolicBP:
        <ul>
            <li>Primer trimestre (0-14): disminución inicial debido a hormonas.</li>
            <li>Segundo trimestre (14-26): alcanza valores más bajos y tiende a estabilizarse.</li>
            <li>Tercer trimestre (27-40): aumenta, sube más que en primeras semanas.</li>
        </ul>
    </li>
    <li>Blood Sugar:
        <ul>
            <li>Primer trimestre: tendencia a hipoglucemia.</li>
            <li>Segundo Trimestre: aumento a resistencia a insulina, elevan niveles de glucosa en sangre.</li>
            <li>Tercer trimestre: se mantiene a nivel del 2do trimestre.</li>
        </ul>
    </li>
    <li>Body Temperature:
        <ul>
            <li>Primer trimestre: suele ser más alta a la normal (0,3-0,4 ºC más, pero se mantiene alrededor de 37 ºC).</li>
            <li>Segundo trimestre: tiende a estabilizarse y regresar a niveles previos al embarazo.</li>
            <li>Tercer trimestre: mantiene niveles 2do trimestre.</li> 
        </ul>
    </li>
    <li>HeartRate:
        <ul>
            <li>Primer trimestre: aumento FC, en reposo es 10 lpm mayor a lo normal</li>
            <li>Segundo trimestre: aumenta FC: incrementa 15 lpm respecto a lo normal,, mayor volumen sanguíneo y mayor retorno venoso. </li>
            <li>Tercer trimestre: pico máximo FC, aumenta 10-20 más en comparación a segundo trimestre.</li> 
        </ul>
    </li>
</ol>


Como observamos, cada variable tiene su patrón fisiológico y modelo, no podemos generalizar y usar un mismotipo de progresión de los cambios para todas, de esta forma, podemos justificar los cambios de cada variable en función de su fisiología. 

Primero, añadimos una columna "week" para ir interpretar los cambios de semana en semana. La fecha de inicio 
será desde la primera consulta a la que acuden y confirman que están embarazadas, y luego al registrarse en la aplicación 
del wearable indican más o menos la fecha que el especialista dijo que quedaron embarazadas. 

```python
#Agregamos columna para calcular semanas
def calculate_weeks(df, user_col = 'user_id', day_col='timestamp', start_date = '2024-06-01 14:38:44'):
    """
    Calcula la semana en la que se encuentra del emebarazo para cada usuario
    ARGS:
        - data (pd.DataFrame): dataframe
        - user_col (str): nombre de la columna con los ids de los usuarios
        - day_col (datetime): nombre de la columna con las fechas
    RETURNS:
        - dataframe con columna adicional "week" (int).
    """
    data = df.copy()
    data[day_col] = pd.to_datetime(data[day_col], errors = 'coerce')
    start_date = pd.to_datetime(start_date)
#calcula las semanas desde el inicio
    data['week'] = ((data[day_col] - start_date).dt.days // 7) +1
    return data

```

**Modelos de simulación**

```python
def stabilize_pattern(start_week, end_week, start_value, stable_value, current_week):
    """
    Simulación de la estabilización  por medio de una función exponencial. 
    Depende de la semana en la que se encuentra, si aún está en el semanas que corresponden 
    al primer trimestre, no se modela nada, en cambio si están en el tercer trimestre, el valor se mantendrá. 
    De lo contrario, se modela cambio. 

    ARGS:
        - start week (int): semana de gestación que empieza  trimestre.
        - end_week (int): última semana de gestación  del trimestre.
        - stable_value (float): valor que se mantiene si está fuera del rtimestre de interés.
        - current_week(int): semana de gestación en la wue se encuentra usuario.
    RETURNS:
        - value (float): valor modificado, estable o de inicio según los casos anteriores.
    
    """
    if current_week < start_week:
        return start_value
    elif current_week > end_week:
        return stable_value
    else:
        k = -np.log(0.1) / (end_week - start_week)
        return start_value + (stable_value - start_value) * (1 - np.exp(-k*(current_week - start_week)))
```

```python
def peak_pattern(start_week, peak_week, end_week, low_value, high_value, current_week):
    """
    Esta función modela cambios en pico, primero  que incrementan/ descienden y luego al contrario. 
    Se usan splines para suavizar la transición. 
    ARGS: 
        - start_week (int): semana de gestación que empieza  trimestre.
        - peak_week (int): semana que alcanza máximo / mínimo valor. 
        - low_value(int): valor mínimo. 
        - high_value(int): valor maximo. 
        - weeks (array): semanas para las  cuales se calculan los cambios.  
    RETURNS:
        - np.array: valores suavizados. 
    """
    spline_weeks = [start_week, peak_week, end_week]
    spline_values = [high_value, low_value, high_value]

    spline_func = UnivariateSpline(spline_weeks, spline_values, k=2, s=0)
    return spline_func(weeks)
        
```

```python
def linear_increase_pattern(start_week, increment_per_week, base_value, current_week):
    """
    Modela cambios lineares. 
    - start_week(int): semana de gestación que empieza  trimestre.
    - increment_per_week(int): incremento a modelar por semana. 
    - base_value(int): valor donde empezamosa modelar. 
    - current week(int): semana actual de gestación. 
    """
    if current_week < start_week:
        return base_value
    else:
        return base_value + increment_per_week*(current_week - start_week)
```

```python
def soft_decreased_pattern(start_week,min_week, end_week, current_value, min_value, current_week):
    """
    Modela fluctuaciones y estabilización de los valores, modela la relación cuadrática: y = a*(x-h)^2+ k, siendo a la curvatura de la parábola.
    siendo h la semana con el valor mínimo, x la semana actual y k el valor mínimo alcanzado. 
    ARGS:
        - start_week(int): semana de gestación que empieza  trimestre.
        - min_week(int): semana donde alcanza valor mínimo. 
        - end_week(int): semana final del trimestre. 
        - low_value(int): valor mínimo. 
        - high_value(int): valor maximo.
        - week(int): semana actual de gestación. 
        
    RETURNS:
        - value (int)
    """
    if current_week < start_week:
        return current_value
    elif current_week > end_week:
        return min_value # valor estabilizado
    else:
       a = (current_value - min_value)/ ((start_week - min_week)**2)
       return a*(current_week - min_week)**2+ min_value
        
```

## VALIDACIÓN DE LOS DATOS

```python
def  adjust_column_formats(df):
    """
    Ajusta las columnas Age, DiastolicBP, SystolicBP, HeartRaten y week como int. 
    ARGS:
        - df: dataframe a modificar. 
    RETURNS:
        - Dataframe con columnas cambiadas.
        
    """
    df['Age'] = df['Age'].astype(int)
    df['week'] = df['week'].astype(int)
    df['DiastolicBP'] = df['DiastolicBP'].astype(int)
    df['SystolicBP'] = df['SystolicBP'].astype(int)
    df['HeartRate'] = df['HeartRate'].astype(int)

    return df
```

```python
from datetime import timedelta
#nos aseguramos que timestamp esté en formato datetime
FT['timestamp'] = pd.to_datetime(FT['timestamp'])

#calcular la diferencia temporal  deseada, desde 02/09/2024 hasta 02/12/2024
shift_days = (pd.Timestamp('2024-09-02') - pd.Timestamp('2024-06-01')).days #extensión temporal que copiamos para el segundo trimestre

#reajustamos


SecondTrimester = FT.copy()
SecondTrimester['timestamp'] = SecondTrimester['timestamp'] + timedelta(days=shift_days)
SecondTrimester['day'] = SecondTrimester['timestamp'].dt.date
SecondTrimester['month&year'] = SecondTrimester['timestamp'].dt.to_period('M').astype(str)
SecondTrimester['date'] = SecondTrimester['timestamp'].dt.date
SecondTrimester = calculate_weeks(SecondTrimester, 'user_id', 'timestamp', '2024-06-01 14:38:44')
SecondTrimester.head()
```

```python
SecondTrimester['DiastolicBP'] = SecondTrimester.apply(
    lambda row: soft_decreased_pattern(14, 20, 26, row['DiastolicBP'], 65, row['week']), axis=1
)

SecondTrimester['SystolicBP'] = SecondTrimester.apply(
    lambda row: soft_decreased_pattern(14, 20, 26, row['SystolicBP'], 110, row['week']), axis=1
)

weeks = SecondTrimester['week']
SecondTrimester['BS'] = peak_pattern(14, 18, 26, 4.2, 13.0, weeks)
SecondTrimester['BodyTemp'] = SecondTrimester['week'].apply(
    lambda x: stabilize_pattern(14, 26, 98.9, 98, x)
)
SecondTrimester['HeartRate'] = SecondTrimester['week'].apply(
    lambda x: linear_increase_pattern(14, 1.5, 100, x)) #para que aumente 5bpm /trimestre, la unción incrementa 1.25 al me
```

```python
def verifica_rango(dataframe):
    
    rangos_limites = {
        'SystolicBP': (70,200),
        'DiastolicBP':(40,120),
        'BodyTemp': (82,106.5),
        'BS':(2.22,33.3),
        'HeartRate':(30,200),
    }
    
    results= {}
    for col, (min_val, max_val) in rangos_limites.items():
        if col in dataframe.columns:
            
            validos = dataframe[(dataframe[col] >= min_val) & (dataframe[col] <= max_val)]
            fuera_rango = len(dataframe) - len(validos)
            results[col] = {
                'val_fuera_rango':fuera_rango,
                'total_filas': len(dataframe), 
                'valores_validos': len(validos)
            }

    for col, info in results.items():
        print(f"- Valores fuera del rango de {col}:{info['val_fuera_rango']}\n")
        print(f"- Total valores de {col}:{info['total_filas']}\n")
        print(f"- Valores validos de {col}: {info['valores_validos']}\n")
    
```

```python
verifica_rango(SecondTrimester)
```
