```python
from scipy.stats import pearsonr
import pandas as pd
import numpy as np

def match_users_directly(df1, df2, columns, correlation_threshold=0.8):
    """
    Correlates every user in df1 with every user in df2 based on the specified columns.
    
    Parameters:
    - df1, df2: DataFrames containing user data.
    - columns: List of column names to correlate.
    - correlation_threshold: Minimum correlation to consider a match.
    
    Returns:
    - DataFrame of user matches with their correlation scores.
    """
    matched_users = []

    # Iterate over each user in df1
    for user1 in df1['user_id'].unique():
        user1_data = df1[df1['user_id'] == user1][columns].dropna().values.flatten()
        if len(user1_data) == 0:
            continue  # Skip users without valid data

        # Iterate over each user in df2
        for user2 in df2['user_id'].unique():
            user2_data = df2[df2['user_id'] == user2][columns].dropna().values.flatten()
            
            # Ensure both arrays have the same length by trimming or skipping
            min_len = min(len(user1_data), len(user2_data))
            if min_len < 2:  # Pearson requires at least two data points
                continue
            
            user1_trimmed = user1_data[:min_len]
            user2_trimmed = user2_data[:min_len]
            
            # Calculate Pearson correlation
            correlation, _ = pearsonr(user1_trimmed, user2_trimmed)
            
            # If correlation exceeds the threshold, consider users matched
            if correlation >= correlation_threshold:
                matched_users.append({
                    'user_id_1': user1,
                    'user_id_2': user2,
                    'correlation': correlation
                })

    # Convert matched users to a DataFrame
    matched_df = pd.DataFrame(matched_users)
    if not matched_df.empty:
        matched_df['unified_id'] = np.arange(1, len(matched_df) + 1)
    return matched_df

# Use this function with the available datasets and columns
common_columns = ['HeartRate', 'SystolicBP', 'DiastolicBP', 'BS']  # Adjust based on actual columns
matched_users_df = match_users_directly(synthea_filled, combined_trimesters, common_columns)

# Display the result or save it
matched_users_df.to_csv('matched_users_direct.csv', index=False)
print(matched_users_df.head())
```

Vamos a quedarnos sólo con las parejas que tengan mayor correlación y vamos a juntar los datasets. 

```python
def get_unique_users(matched_users_df):
    """
    Extrae pares únicos de usuarios con mayor correlación. 
    ARGS: 
        - matched_users_df : dataframe con las columnas ['user_id_1', 'user_id_2', 'correlation']
    RETURNS:
        - Dataframe con pares únicos de usuarios mayormente correlacionados.
    
    """
    #ordenamos en orden descendente la correlacion
    sorted_df = matched_users_df.sort_values('correlation', ascending = False)
    matched_user_1 = set()
    matched_user_2 = set()
    unique_pairs = []

    #iteramos para elegir las parejas
    for _, row in sorted_df.iterrows():
        user1, user2 = row['user_id_1'], row['user_id_2']
        if user1 not in matched_user_1 and user2 not in matched_user_2:
            unique_pairs.append(row)
            matched_user_1.add(user1)
            matched_user_2.add(user2)
    #convertimos en dataFrame
    return pd.DataFrame(unique_pairs)
            
```

**Alineación de las fecha de EncounterData con WearableData**

Vamos a mapear las fechas deusuarios en ambos datasets, para cada usuario en el dataset de <code>synthea_filled</code>, 
asignaremos ua fecha base ajustada a las del dataset de <code>combined_trimesters</code>. 

Tras ello, ajustaremos las fechas en synthea_filled, ya que sin datos sintéticos es muy representativo y podemos modularlos, elproblema sería
si fueran datos reales. Sin embargo, el intervalo entre consultas para un mismo usuario se va a mantener, ya que este patrón es importante para documentar el progreso entre visitas y la diferenciacia de los datos entre el sistema propuesto y los datos de la consulta,  la línea de investigación del presente Trabajo de Final de Carrera se basa en esto. 

```python
#creamos un diccionario de mapeo desde unique_matched_users
#common_id conect los IDs de synthea_filled (user_id) con los del wearable (user_XXXX)

user_id_mapping = unique_matched_users.set_index('user_id_1')['common_id'].to_dict()

#Agregar columna de IDs mapeados al dataframe synthea_filled
synthea_filled['common_id'] = synthea_filled['user_id'].map(user_id_mapping)
synthea_filled[['user_id', 'common_id']].head()
```

```python
common_wearable_mapping = unique_matched_users.set_index('common_id')['user_id_2'].to_dict()

synthea_filled['wearable_user_id'] = synthea_filled['common_id'].map(common_wearable_mapping)
print("Wearable user IDs en synthea_filled:", synthea_filled['wearable_user_id'].unique())
```

```python
#Nos aseguramos que las fecha estén en formato datetime

synthea_filled['DATE'] = pd.to_datetime(synthea_filled['DATE'])
combined_trimesters['day'] = pd.to_datetime(combined_trimesters['day'])

#Creamos un dicconario con las fecha más temprana (start_date) por usuario en el dataset combined_trimesters
wearables_start_dates = combined_trimesters.groupby('user_id')['day'].min().to_dict()


#columna auxiliar para asignar el número  de consulta por usuario
synthea_filled['visit_number'] = synthea_filled.groupby('user_id').cumcount() + 1
```

```python
def adjust_date(row, start_dates):
    """
    Función para ajustar las fechas manteniendo el intervalo relativo.
    ARGS:
        - row:  fila con user_id. 
        - start_dates: primera fecha de la visita de ese usuario. 
    RETURNS:
        - Fecha ajustada si no está ya modificada.
    """
    base_date = start_dates.get(row['wearable_user_id'], None)
    if base_date is not None:
        min_date = synthea_filled.loc[
        synthea_filled['wearable_user_id'] == row['wearable_user_id'], 'DATE'
        ].min()
        interval = (row['DATE'] - min_date).days #diferencia en días
        return base_date + timedelta(days=interval)
    return None
    
```

```python
synthea_filled['Adjusted_DATE'] = synthea_filled.apply(
    lambda row: adjust_date(row, wearables_start_dates), axis=1
)
synthea_filled[['user_id', 'common_id', 'wearable_user_id',  'visit_number', 'DATE', 'Adjusted_DATE']].head()
```
