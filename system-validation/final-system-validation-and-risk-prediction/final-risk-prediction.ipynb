```python
import pandas as pd

def normalize_columns(df, id_col, risk_col, date_col=None):
    df = df.rename(columns={id_col: 'PATIENT', risk_col: 'Predicted_Risk'})
    if date_col and date_col in df.columns:
        df[date_col] = pd.to_datetime(df[date_col])
        df[date_col] = df[date_col].apply(lambda x: x.replace(year=2024))  # Cambiar el año a 2024
        df['MONTH'] = df[date_col].dt.to_period('M')
    return df


def combine_predictions(consultation_df, wearable_df, weight_consultation=0.7, weight_wearable=0.3, preeclampsia_df=None, gdm_df=None):
    """
    Combina las predicciones de consulta y wearable, dando más peso a la consulta.
    Si la consulta indica alto riesgo de preeclampsia o GDM, el riesgo maternal se ajusta.
    """
    # Normalizar columnas
    consultation_df = normalize_columns(consultation_df, 'PATIENT', 'Predicted_Risk')
    wearable_df = normalize_columns(wearable_df, 'PATIENT', 'Predicted_Risk', 'DATE')
    
    # Fusionar con inner join para mantener solo pacientes comunes
    merged_df = consultation_df.merge(
        wearable_df, on=['PATIENT'], suffixes=('_consultation', '_wearable'), how='inner'
    )
    
    # Si hay dataset de preeclampsia, fusionarlo
    if preeclampsia_df is not None:
        preeclampsia_df = normalize_columns(preeclampsia_df, 'PATIENT', 'Predicted_Risk')
        merged_df = merged_df.merge(preeclampsia_df[['PATIENT', 'Predicted_Risk']], on='PATIENT', how='left', suffixes=('', '_preeclampsia'))
        merged_df['Predicted_Risk_preeclampsia'].fillna(0, inplace=True)
    
    # Si hay dataset de GDM, fusionarlo
    if gdm_df is not None:
        gdm_df = normalize_columns(gdm_df, 'PATIENT', 'Predicted_Risk')
        merged_df = merged_df.merge(gdm_df[['PATIENT', 'Predicted_Risk']], on='PATIENT', how='left', suffixes=('', '_gdm'))
        merged_df['Predicted_Risk_gdm'].fillna(0, inplace=True)
    
    # Verificar si las columnas existen antes de usarlas
    if 'MONTH_wearable' in merged_df.columns:
        merged_df['MONTH'] = merged_df['MONTH_wearable']
    
    if 'Predicted_Risk_consultation' not in merged_df.columns:
        merged_df['Predicted_Risk_consultation'] = 0
    if 'Predicted_Risk_wearable' not in merged_df.columns:
        merged_df['Predicted_Risk_wearable'] = 0
    
    # Asignar valores por defecto si hay datos faltantes
    merged_df['Predicted_Risk_consultation'].fillna(0, inplace=True)
    merged_df['Predicted_Risk_wearable'].fillna(0, inplace=True)
    
    # Calcular predicción final ponderada y convertir a entero
    merged_df['Final_Prediction'] = (
        merged_df['Predicted_Risk_consultation'] * weight_consultation +
        merged_df['Predicted_Risk_wearable'] * weight_wearable
    ).round().astype(int)
    
    # Ajustar riesgo maternal si preeclampsia o GDM es alta
    if preeclampsia_df is not None:
        merged_df.loc[merged_df['Predicted_Risk_preeclampsia'] >= 2, 'Final_Prediction'] = \
            merged_df['Final_Prediction'].clip(lower=1)
    
    if gdm_df is not None:
        merged_df.loc[merged_df['Predicted_Risk_gdm'] >= 2, 'Final_Prediction'] = \
            merged_df['Final_Prediction'].clip(lower=1)
    # Seleccionar columnas relevantes
    final_df = merged_df[['PATIENT', 'MONTH', 'Predicted_Risk_consultation', 'Predicted_Risk_wearable', 'Final_Prediction']]
    
    return final_df


```

```python
encounterMaternal = normalize_columns(encounterMaternal, 'user_id', 'Predicted_Risk',  'day')
wearableMaternal = normalize_columns(wearableMaternal, 'PATIENT', 'Predicted_Risk', 'DATE')
encounterGdm = normalize_columns(encounterGdm, 'PATIENT ID', 'Risk prediction', None)
wearableGdm = normalize_columns(wearableGdm, 'PATIENT', 'Predicted Target', 'DATE')
wearablePreeclampsia = normalize_columns(wearablePreeclampsia, 'PATIENT', 'Predicted Target', 'DATE')
encounterPreeclampsia =normalize_columns(encounterPreeclampsia, 'PATIENT', 'Predicted_Class', None)
```

```python
def plot_patient_risk(final_df, patient_id, ax=None, title="Evolución del Riesgo"):
    """
    Genera una gráfica de la evolución del riesgo de un paciente a lo largo del tiempo.
    """
    patient_data = final_df[final_df['PATIENT'] == patient_id].sort_values(by='MONTH')
    
    if patient_data.empty:
        print(f"No hay datos disponibles para el paciente {patient_id}")
        return
    
    patient_data['MONTH'] = patient_data['MONTH'].astype(str)  # Asegurar que el eje X sea correcto
    
    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 5))
    
    ax.plot(patient_data['MONTH'], patient_data['Predicted_Risk_consultation'], marker='o', linestyle='--', label='Consulta')
    ax.plot(patient_data['MONTH'], patient_data['Predicted_Risk_wearable'], marker='s', linestyle='--', label='Wearable')
    ax.plot(patient_data['MONTH'], patient_data['Final_Prediction'], marker='d', linestyle='-', label='Riesgo Final')
    
    ax.set_ylabel('Nivel de Riesgo')
    ax.set_title(title)
    ax.legend()
    ax.grid()
```

```python
plot_patient_risk(maternalPred, '0ab45617-d629-d799-de91-c18ef91fd058', None, "Evolución del Riesgo Maternal")
```

```python
def plot_all_risks(maternalPred, gdmPred, preeclampsiaPred, patient_id):
    """
    Genera un grid de subplots con la evolución de los tres riesgos para un paciente.
    """
    fig, axes = plt.subplots(3, 1, figsize=(10, 15), sharex=True)
    
    plot_patient_risk(maternalPred, patient_id, axes[0], 'Evolución del Riesgo Maternal')
    plot_patient_risk(gdmPred, patient_id, axes[1], 'Evolución del Riesgo de GDM')
    plot_patient_risk(preeclampsiaPred, patient_id, axes[2], 'Evolución del Riesgo de Preeclampsia')
    
    axes[-1].set_xlabel('Mes')
    plt.xticks(rotation=45)
    plt.show()

```
